# analysis.ipynb
This notebook contains the main code for Deliverable 1 (D1): Data Analysis.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr, ttest_1samp

# Load dataset
dataset = pd.read_csv('StudentPerformanceFactors.csv')
dataset.head()
```

## Descriptive Statistics


```python
dataset.describe()
```

## Random Sampling


```python
sample_data = dataset.sample(n=150, random_state=42)
sample_data.head()
```

## Systematic Sampling


```python
step = 10
systematic_sample = dataset.iloc[::step, :]
systematic_sample['Exam_Score'].describe()
```

## Scatter Plot


```python
plt.figure(figsize=(8,5))
plt.scatter(sample_data['Hours_Studied'], sample_data['Exam_Score'])
plt.title('Scatter Plot: Hours_Studied vs Exam_Score')
plt.xlabel('Hours_Studied')
plt.ylabel('Exam_Score')
plt.show()
```

## Box Plot


```python
plt.figure(figsize=(8,5))
sns.boxplot(x=dataset['Exam_Score'])
plt.title('Box Plot of Student Exam Scores')
plt.show()
```

## Histogram


```python
plt.figure(figsize=(8,5))
sns.histplot(dataset['Exam_Score'], bins=10)
plt.title('Histogram of Student Exam Scores')
plt.xlabel('Exam Score')
plt.ylabel('Frequency')
plt.show()
```

## Correlation Heatmap


```python
numeric_df = dataset.select_dtypes(include=['number'])
corr = numeric_df.corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='Blues', linewidths=0.5)
plt.title('Heat Map of Correlation between Student Performance Factors')
plt.show()
```

## Pearson Correlation Test


```python
hours = dataset['Hours_Studied'].dropna()
exam_score = dataset['Exam_Score'].dropna()
min_len = min(len(hours), len(exam_score))
hours = hours[:min_len]
exam_score = exam_score[:min_len]

r, p = pearsonr(hours, exam_score)
print('Pearson Correlation (r):', round(r, 3))
print('P-value:', round(p, 3))
```

## One Sample T-Test


```python
t_stat, p_value = ttest_1samp(dataset['Exam_Score'].dropna(), 70)
print('T-statistic:', round(t_stat, 3))
print('P-value:', round(p_value, 3))
```
